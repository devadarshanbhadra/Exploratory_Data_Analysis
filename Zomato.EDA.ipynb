## Dataset country code:?

```python
df_country=pd.read_excel('Country-Code.xlsx')          #if not showed then encoding is done
df_country.head()
```

Thus country code is added to csv file using merge command

```python
pd.merge(df,df_country,on='Country Code', how='left')
```

```python
final_df = pd.merge(df,df_country,on='Country Code', how='left')
```

```python
final_df.head(2)
```

```python
final_df.columns
```

Now taking a feature and applying observations on it

```python
final_df.Country.value_counts()
```

```python
country_names=final_df.Country.value_counts().index
```

```python
country_val=final_df.Country.value_counts().values
```

### Plotting some pie chart comparing countries using zomato
Using plt.pie

```python
## pie chart - Tp 3 countries uses zomato
## giving percentage functions 
plt.pie(country_val[:3],labels=country_names[:3],autopct='%1.2f%%')
```

Observations:
1. In India zomato is using maximum
2. After that usa and uk is using zomato

O is renamed as Rating count

```python
final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating count'})
```

```python
Ratings = final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating count'})
```

## Making some visualisation and ploting on this ratings:

## Observation:
1. When rating is 4.5-4.9 ---> Excelent(Dark green)
2. When Rating is 4.0 - 4.4 --->Very good(green)
3. When Rting is 3.5-3.9 --> good(Yellow)
4. When Rating Is 2.5-3.4--> Averrage(orange)
5. When Rating is 0-2.4 --> poor(Red)*italicized text*

```python
Ratings.head()
```

# Visualisation using sea born

```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,6)
sns.barplot(x='Aggregate rating',y='Rating count',data = Ratings)
```

Applying colours to visuualisation using hue command

```python
sns.barplot(x='Aggregate rating',y='Rating count',hue ='Rating color' ,data = Ratings)
```

Mapping the colours accoding as per data in the list:

```python
sns.barplot(x='Aggregate rating',y='Rating count',hue ='Rating color' ,data = Ratings,palette=['white','red','orange','yellow','green','green'])
```

## Observation:
 1. Not Rated is very high.
 2. Maxumum number of ratings is very high between 2.5-3.4

### Count PLot-Used for plotting Categorical variables

```python
sns.countplot(x='Rating color',data = Ratings,palette=['white','red','orange','yellow','green','green'])
```
